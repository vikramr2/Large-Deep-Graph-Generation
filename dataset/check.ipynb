```python
import pickle

file_path = "community.pkl"

with open(file_path, "rb") as file:
    data = pickle.load(file)

if isinstance(data, dict):
    print("Keys:", data.keys())
    for key, value in data.items():
        print(f"{key}: {type(value)}")
elif isinstance(data, list):
    print(f"List length: {len(data)}")
    print(f"First element type: {type(data[0])}")
else:
    print(f"Data type: {type(data)}")

try:
    import networkx as nx
    if isinstance(data, nx.Graph):
        print("Graph info:", nx.info(data))
    else:
        print("Dataset contains other types.")
except ImportError:
    pass
```

```python
import pickle
import networkx as nx

file_path = "community.pkl"
with open(file_path, "rb") as file:
    data = pickle.load(file)

print(f"Dataset type: {type(data)}")
print(f"Number of graphs in dataset: {len(data)}")
print(f"Type of first element: {type(data[0])}")

graph = data[1]

print("Graph Info:")
print(f"Number of nodes: {graph.number_of_nodes()}")
print(f"Number of edges: {graph.number_of_edges()}")

print("\nNodes:")
print(graph.nodes(data=True))
print("\nEdges:")
print(graph.edges(data=True))

try:
    import matplotlib.pyplot as plt

    nx.draw(graph, with_labels=True, node_size=500, node_color="lightblue", font_size=8)
    plt.show()
except ImportError:
    print("Matplotlib is not installed.")
```

```python
graph = data[1]

print("Nodes with attributes:")
for node, attr in graph.nodes(data=True):
    print(f"Node {node}: {attr}")

print("\nEdges with attributes:")
for u, v, attr in graph.edges(data=True):
    print(f"Edge ({u}, {v}): {attr}")

print("\nGraph attributes:", graph.graph)
```
