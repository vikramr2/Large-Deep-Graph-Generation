```python
import pickle
import networkx as nx

# Load the dataset
file_path = "ENZYMES.pkl"
with open(file_path, "rb") as file:
    data = pickle.load(file)

# Check the structure
print(f"Dataset type: {type(data)}")
print(f"Number of graphs in dataset: {len(data)}")
print(f"Type of first element: {type(data[0])}")

# Inspect the first graph
graph = data[1]  # Get the first graph

# Print basic information
print("Graph Info:")
print(f"Number of nodes: {graph.number_of_nodes()}")
print(f"Number of edges: {graph.number_of_edges()}")

# Print nodes and edges
print("\nNodes:")
print(graph.nodes(data=True))  # Node attributes
print("\nEdges:")
print(graph.edges(data=True))  # Edge attributes

# Visualize the graph (optional)
try:
    import matplotlib.pyplot as plt

    nx.draw(graph, with_labels=True, node_size=500, node_color="lightblue", font_size=8)
    plt.show()
except ImportError:
    print("Matplotlib is not installed.")
```

```python
import os
import pickle
import networkx as nx

folder_path = "/content/drive/MyDrive/cit-HepPh"

graphs = []

for file_name in sorted(os.listdir(folder_path)):
    if file_name.endswith(".tsv"):
        file_path = os.path.join(folder_path, file_name)

        # Create a new graph for each file
        graph = nx.Graph()

        with open(file_path, "r") as file:
            for line in file:
                node1, node2 = map(int, line.strip().split("\t"))  # Parse nodes from the line
                graph.add_edge(node1, node2, label='DEFAULT_LABEL')  # Add edge with 'label'

        # Add the 'label': 'DEFAULT_LABEL' attribute to all nodes
        nx.set_node_attributes(graph, 'DEFAULT_LABEL', name='label')

        # Append the graph to the list
        graphs.append(graph)

output_path = "/content/drive/MyDrive/cit-HepPh.pkl"
with open(output_path, "wb") as pkl_file:
    pickle.dump(graphs, pkl_file)

print(f"Serialized {len(graphs)} graphs to {output_path}")
```
